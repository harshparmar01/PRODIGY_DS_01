## Removing Unnecessary Columns

```python
df.drop(['Series Name','Country Name'],axis=1,inplace=True)
```

## Top 10 Countries with respect to Total Population

```python
#filtering the data
tot_population = df[df['Series Code'] == 'SP.POP.TOTL']
#sort the data based on total population of 2022
tot_population_sorted = tot_population.sort_values(by="2022",ascending=False)
#get the top ten countries with highest total population for 2022
total_top_ten_countries = tot_population_sorted.head(10)

print("Top 10 Countries of total population: ")
print(total_top_ten_countries[['Country Code']])
```

## Bar Plotting Top 10 countries of total population in year 2022 and 2016

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data= total_top_ten_countries,palette="viridis")
plt.title("Top 10 countries of total population of 2022",fontsize=12)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2020", y="Country Code", data= total_top_ten_countries,palette="viridis")
plt.title("Top 10 countries of total population of 2020",fontsize=12)
plt.xlabel("Total Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```

## Extracting top 10 countries with highest male population and highest female population

```python
#filtering data for male population
male_population = df[df['Series Code'] == 'SP.POP.TOTL.MA.IN']
#sort the data based on total male population of 2022
tot_male_population_sorted = male_population.sort_values(by="2022",ascending=False)
#get the top ten countries with highest male population for 2022
male_top_ten_countries = tot_male_population_sorted.head(10)

print("Top 10 Countries with highest male population: ")
print(male_top_ten_countries[['Country Code']])
```

```python
#filtering data for female population
female_population = df[df['Series Code'] == 'SP.POP.TOTL.FE.IN']
#sort the data based on total female population of 2022
tot_female_population_sorted = female_population.sort_values(by="2022",ascending=False)
#get the top ten countries with highest female population for 2022
female_top_ten_countries = tot_female_population_sorted.head(10)

print("Top 10 Countries with highest female population: ")
print(female_top_ten_countries[['Country Code']])
```

## Bar Plotting of top 10 countries with highest male and female in 2022

```python
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Code", data= male_top_ten_countries,palette="coolwarm")
plt.title("Top 10 countries having highest male population of 2022",fontsize=12)
plt.xlabel("Male Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()

plt.figure(figsize=(15,6))
plt.subplot(2,2,2)
sns.barplot(x="2020", y="Country Code", data= female_top_ten_countries,palette="coolwarm")
plt.title("Top 10 countries having highest female population of 2022",fontsize=12)
plt.xlabel("Female Population",fontsize=10)
plt.ylabel("Country",fontsize=10)
plt.show()
```
